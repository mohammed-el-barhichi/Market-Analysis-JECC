```python
import matplotlib.pyplot as plt
import pandas as pd


# import data from xlsx file
df = pd.read_excel('Choix.xlsx')
df['Quantitée choisie'] = df['Quantitée choisie'].str.lower()
df
```

```python
df['Quantitée choisie'] = df['Quantitée choisie'].str.strip()

# create a copy of the dataframe
df_choix = df.copy()

df1 = df_choix[df_choix['Quantitée choisie'].str.startswith('1')]
df2 = df_choix[df_choix['Quantitée choisie'].str.startswith('7')]
df3 = df_choix[df_choix['Quantitée choisie'].str.startswith('l')]
df4 = df_choix[df['Quantitée choisie'].str.startswith('1') == False][df['Quantitée choisie'].str.startswith('7') == False][df['Quantitée choisie'].str.startswith('l') == False]
```

```python
# replace all values in the column 'Quantitée choisie' that start with '1' by '1'
df1['Quantitée choisie'] = '1 kg'
df2['Quantitée choisie'] = '750 g'
df3['Quantitée choisie'] = 'Les deux choix'
df4['Quantitée choisie'] = 'Aucun choix'
```

```python
# concat all dataframes
df_choix = pd.concat([df1, df2, df3, df4])
df_choix
```

```python
# plot the data in a pie chart using matplotlib library
df_choix['Quantitée choisie'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(10, 10))
plt.title('Choix des intervenants')
plt.ylabel('')
plt.show()
```

```python
# plot bar chart
df_choix['Quantitée choisie'].value_counts().plot(kind='bar', figsize=(10, 10))
plt.title('Choix des intervenants')
plt.ylabel('Nombre de personnes')
plt.show()
```

```python
# import data from xlsx file sheet 2
df = pd.read_excel('Choix.xlsx', sheet_name='Feuil3')
df['Quantitée choisie'] = df['Quantitée choisie'].str.lower()
df['Quantitée choisie'] = df['Quantitée choisie'].str.strip()
df
```

```python
df1 = df[df['Quantitée choisie'].str.startswith('1')]
df2 = df[df['Quantitée choisie'].str.startswith('7')]
df3 = df[df['Quantitée choisie'].str.startswith('l')]
df4 = df[df['Quantitée choisie'].str.startswith('1') == False][df['Quantitée choisie'].str.startswith('7') == False][df['Quantitée choisie'].str.startswith('l') == False]
```

```python
# remplacer "1kg" par "" dans des lignes de df1
df1['Quantitée choisie'] = df1['Quantitée choisie'].str.replace('1kg', '')
df1['Quantitée choisie'] = df1['Quantitée choisie'].str.replace('1 kg', '')
df1['Quantitée choisie'] = df1['Quantitée choisie'].str.replace('1', '')
df1['Quantitée choisie'] = df1['Quantitée choisie'].str.replace('raison', '')
df1
```

```python
# remplacer "750g" par "" dans des lignes de df2
df2['Quantitée choisie'] = df2['Quantitée choisie'].str.replace('750g', '')
df2['Quantitée choisie'] = df2['Quantitée choisie'].str.replace('750 g', '')
df2['Quantitée choisie'] = df2['Quantitée choisie'].str.replace('750 kg', '')
df2['Quantitée choisie'] = df2['Quantitée choisie'].str.replace('750kg', '')
df2['Quantitée choisie'] = df2['Quantitée choisie'].str.replace('750', '')
df2['Quantitée choisie'] = df2['Quantitée choisie'].str.replace('raison', '')
df2
```

```python
# remplacer "les deux" par "" dans des lignes de df3
df3['Quantitée choisie'] = df3['Quantitée choisie'].str.replace('les deux', '')
df3['Quantitée choisie'] = df3['Quantitée choisie'].str.replace('raison', '')
df3
```

```python
from wordcloud import WordCloud, STOPWORDS

stopwords = set(STOPWORDS)
feedback_text = ' '.join(df1['Quantitée choisie'])
```

```python
wc = WordCloud(width=800, height=400, background_color='white', min_font_size=10)
wc.generate(feedback_text)
plt.figure(figsize=(14,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
